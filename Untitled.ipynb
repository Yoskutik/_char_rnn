```python
import pandas as pd
import requests
import random
import time
import glob
from bs4 import BeautifulSoup
from awesome_progress_bar import ProgressBar
```

```python
padeji = {
    'им': 0,
}

def read(filename):
    a, b, c, d = filename[6:].replace('.txt', '').split('_')
    data = {}
    with open(filename, 'rb') as f:
        lines = f.readlines()[2:]
        data['Слово'] = [x.decode('utf-8').rstrip('\r\n') for x in lines]
        data['Часть речи'] = [a] * len(lines)
        data['Род'] = [b] * len(lines)
        data['Число'] = [c] * len(lines)
        data['Падеж'] = [padeji[d]] * len(lines)
        data['Ударение'] = [None] * len(lines)
    return data
```

```python
def read_all():
    df = pd.DataFrame([], columns=['Слово', 'Часть речи', 'Род', 'Число', 'Падеж', 'Ударение'])
    for path in glob.glob('words/*.txt'):
        data = read(path)
        df = df.append(pd.DataFrame.from_dict(data))
    return df.reset_index()
```

```python
df = read_all()
df.shape
```

```python
vowels = 'а у о ы и э я ю ё е'.split(' ')
```

```python
sites = [
    'https://accentonline.ru',
    'https://wordsonline.ru',
    'https://phoneticonline.ru',
    'https://xn----8sbhebeda0a3c5a7a.xn--p1ai',
]

def get_strike_idx(word):
    if not any([l in vowels for l in word]):
        return -1
    l = word[0]
    tmp_sites = sites[:]
    random.shuffle(tmp_sites)
    for i in range(len(tmp_sites)):
        site = tmp_sites[i]
        print(site)
        try:
            if site == 'https://accentonline.ru':
                soup = BeautifulSoup(requests.get(f'{site}/{l}/{word}').content)
                return soup.select_one('.word-accent').text.index('\u0301')
            elif site == 'https://wordsonline.ru':
                soup = BeautifulSoup(requests.get(f'{site}/{l}/{word}').content)
                return soup.select_one('.orthography').text.index('\u0301')
            elif site == 'https://phoneticonline.ru':
                soup = BeautifulSoup(requests.get(f'{site}/{word}').content)
                return soup.select_one('.word-accent').text.index('\u0301')
            elif site == 'https://xn----8sbhebeda0a3c5a7a.xn--p1ai':
                soup = BeautifulSoup(requests.get(f'{site}/в-слове-{word}').content)
                return soup.select('b')[2].text.index('\u0301')
        except:
            pass
    return None
```

```python
for i in range(df.shape[0]):
    if df.loc[i, 'Ударение']: continue
    print(i, end='')
    time.sleep(0.3)
    idx = get_strike_idx(df.loc[i, 'Слово'])
    df.loc[i, 'Ударение'] = idx
```
